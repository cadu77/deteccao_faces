```python
imagem = cv2.resize(imagem, (800, 600))
imagem.shape
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
```

```python
imagem_cinza.shape
```

# **Detectando as faces**

```python
detector_facil = cv2.CascadeClassifier('/content/drive/MyDrive/Visão Computacional/Cascades/haarcascade_frontalface_default.xml')
```

```python
deteccoes = detector_facil.detectMultiScale(imagem_cinza)
```

```python

deteccoes
```

```python
len(deteccoes)
```

```python
for x, y, w, h in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0,255,255), 5)
cv2_imshow(imagem)
```
